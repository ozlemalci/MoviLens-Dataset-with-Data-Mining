```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

baslik= ['KullaniciId', 'FilmId','Degerlendirme', 'ZamanDamgasi']
df= pd.read_csv('ml-100k/ml-100k/u.data', sep='\t',names= baslik)

KullaniciSayisi=df.KullaniciId.unique().shape[0]
FilmSayisi=df.FilmId.unique().shape[0]

print('Kullanıcı Sayısı = ' +str(KullaniciSayisi)+', Film Sayısı =' +str(FilmSayisi))

df.head()

```

```
plt.rc("font", size=10)
df.Degerlendirme.value_counts(sort=False).plot(kind ="bar")
plt.title("Değerlendirme Dağılımı")
plt.xlabel("Değer")
plt.ylabel("Beğeni Sayısı")
plt.show()
```
